## $\texttt{Part One}$

这一部分自然是关于微分方程的，有两道大题（代码题），一道是ODE，考察就是给定了一个弹簧的方程：

$$
m \ddot{r} + C \dot{r} + k r = F_0 \cos(\omega t)
$$

要求写出`def dy_dy(x,*args)`。以及最后有关于CTE步长和精度的分析。

另一道则是PDE，1d波动方程。这一道比较诡异，要求通过调整边界条件搞出“有5个节点”“有三个节点”的驻波。还需要设置动画演示。


## $\texttt{Part Two}$

这一部分是关于Monte Carlo的，第一道是拒绝采样，但是Tricky的点在于，给出的$p(x)$并没有归一化，但最后会要求Plot归一化的图。所以该题的第一步是先通过MC方法或是`quad`求出归一化的分母，之后可以正常进行。

第二题则是MH算法，该题较为常规，给了一个奇怪一点的目标分布，改改公式即可。

## $\texttt{Part Three}$

该部分是一道送分代码题+16道选择题。送分的代码题为使用`LogisticRegression,SVM,MLP`在Digits数据集上做分类，要求使用8-Fold交叉验证，最后Print Score。

16道选择题，则类似智慧树线上课程的类型。
